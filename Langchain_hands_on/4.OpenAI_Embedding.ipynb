#### TEXT LOADER

```python
## TEXT LOADER
from langchain_community.document_loaders import TextLoader

# Intializer loader 
loader = TextLoader("/home/mandeep/Desktop/mandeep_personal/Agentic-AI/Data_sources/speech.txt")
loader
```

```python
docs = loader.load()
docs
```

```python
from langchain_text_splitters import RecursiveCharacterTextSplitter

text_spliiter = RecursiveCharacterTextSplitter(chunk_size = 100 , chunk_overlap = 20)
text_spliiter
```

```python
final_documents = text_spliiter.split_documents(docs)
final_documents
```

```python
# Vector Embedding and Vector Store DB
from langchain_community.vectorstores import Chroma

db = Chroma.from_documents(final_documents, embeddings)
db
```

```python
query = "It will be all the easier for us to conduct ourselves as belligerents in a high spirit of right"
retrieved_response = db.similarity_search(query)
print(retrieved_response)
```
